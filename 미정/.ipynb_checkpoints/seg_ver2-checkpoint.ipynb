고객 groupby : 구매 물품 종류, *가입기간 별* (12개월 미만 / 12개월 / 24개월 / 36개월 / 48개월 / 50개월)

```python
customer_df
```

```python
merged_df
```

```python
def segmentation(x):
    if x >= 50 :
        return "vip"
    elif x >= 48 : 
        return "diamond"
    elif x >= 36 :
        return "gold"
    elif x >= 24 :
        return "silver"
    elif x >= 12 : 
        return "bronze"
    else :
        return "new"
```

```python
merged_df['고객등급'] = merged_df['가입기간'].apply(segmentation)
```

```python
merged_df
```

```python
pd.DataFrame(merged_df.groupby(["고객등급", "제품카테고리"])['수량'].sum())
```

```python
# 데이터프레임 생성
df = pd.DataFrame(merged_df.groupby(["고객등급", "제품카테고리"])['수량'].sum()).reset_index()

# Bar 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='제품카테고리', y='수량', hue='고객등급', data=df)
plt.xticks(rotation=90, fontsize=8)
plt.title('고객등급 및 제품카테고리별 수량')
plt.show()
```

```python
# 고객등급이 'vip'인 행을 필터링한 후 제품카테고리로 그룹화하여 수량 합계 구하기
vip_category_sum = merged_df[merged_df['고객등급'] == 'vip'].groupby('제품카테고리')['수량'].sum()

# 수량 합계를 기준으로 내림차순 정렬
vip_category_sum = vip_category_sum.sort_values(ascending=False)

# seaborn의 barplot으로 바 그래프 그리기
ax = sns.barplot(x=vip_category_sum.index, y=vip_category_sum.values)

# 바 위에 수치 표시
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# 그래프에 제목과 축 레이블 추가
plt.title('VIP 고객의 제품 카테고리별 수량 합계')
plt.xlabel('제품 카테고리')
plt.ylabel('수량')

# x축 레이블 회전
plt.xticks(rotation=45, ha='right')

# 그래프 표시
plt.show()
```
