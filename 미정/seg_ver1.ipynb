```python
import folium
from folium.plugins import MarkerCluster
```

```python
# 1. folium으로 시각화 
```

```python
# 데이터 프레임 생성
data = {
    '고객수': [16136, 18380, 4503, 11173, 2732],
    '서경': [-120.0, -87.62777778, -74.5, -74.00611111, -77.03666667],
    '북위': [37.0, 41.88194444, 40.0, 40.71277778, 38.895],
    '고객지역': ['California', 'Chicago', 'New Jersey', 'New York', 'Washington DC']
}

df = pd.DataFrame(data)

# 위도 경도 값으로 Folium 맵 생성
m = folium.Map(location=[37.7749, -122.4194], zoom_start=4)  # 기본 위치는 미국 중심

# 각 지역에 원형 마커 추가
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['북위'], row['서경']],
        radius=row['고객수'] / 500,  # 반지름을 고객 수에 비례하게 조절
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.7,
        tooltip=f"{row['고객지역']}: {row['고객수']} 명"
    ).add_to(m)

m
```
