고객 groupby : 구매 물품 종류, *가입기간 별* (12개월 미만 / 12개월 / 24개월 / 36개월 / 48개월 / 50개월)

```python
customer_df
```

```python
merged_df
```

```python
def segmentation(x):
    if x >= 50 :
        return "vip"
    elif x >= 48 : 
        return "diamond"
    elif x >= 36 :
        return "gold"
    elif x >= 24 :
        return "silver"
    elif x >= 12 : 
        return "bronze"
    else :
        return "new"
```

```python
merged_df['고객등급'] = merged_df['가입기간'].apply(segmentation)
```

```python
merged_df
```

```python
pd.DataFrame(merged_df.groupby(["고객등급", "제품카테고리"])['수량'].sum())
```

```python
%matplotlib inline
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터프레임 생성
df = pd.DataFrame(merged_df.groupby(["고객등급", "제품카테고리"])['수량'].sum()).reset_index()

# Bar 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='제품카테고리', y='수량', hue='고객등급', data=df)
plt.xticks(rotation=90, fontsize=8)
plt.title('고객등급 및 제품카테고리별 수량')
plt.show()
```
