```python
customer_df = customer_df.sort_values(by='고객ID').reset_index(drop=True)
customer_df
```

```python
sales_df.groupby('고객ID')['거래날짜'].max()
```

```python
sales_df.groupby('고객ID')['거래날짜'].max().reset_index()['거래날짜']
```

```python
customer_df['recency_date'] = sales_df.groupby('고객ID')['거래날짜'].max().reset_index()['거래날짜']
```

```python
customer_df
```

```python
# 날짜 형식으로 변경
customer_df['recency_date'] = pd.to_datetime(customer_df['recency_date'])
```

```python
customer_df.info()
```

```python
customer_df['recency_diff'] = (pd.to_datetime('2020-01-01') - customer_df['recency_date']).dt.days
```

```python
customer_df
```

```python
# frequency는 거래 날짜로 묶어야 할 것 같음 
# 같은 날 여러번 들어 온 것은 재방문으로 보기 어렵다고 생각 -> 날짜가 다른 경우만 카운트

frequency_df = pd.DataFrame(sales_df.groupby(['고객ID', '거래날짜']).nunique()).reset_index()
frequency_df
```

```python
# 거래날짜를 카운팅해서 몇 번 방문했는지 계산

pd.DataFrame(frequency_df.groupby('고객ID')['거래날짜'].count()).reset_index()['거래날짜']
```

```python
customer_df['frequency'] = pd.DataFrame(frequency_df.groupby('고객ID')['거래날짜'].count()).reset_index()['거래날짜']
customer_df
```

```python
sales_df['금액'] = sales_df['평균금액'] * sales_df['수량']
sales_df
```

```python
sales_df.groupby('고객ID')['금액'].sum()
```

```python
sales_df.groupby('고객ID')['금액'].sum().reset_index()['금액']
```

```python
customer_df['monetary'] = sales_df.groupby('고객ID')['금액'].sum().reset_index()['금액']
customer_df
```

```python
pd.qcut(customer_df['recency_diff'], q=5)
```

```python
pd.cut(customer_df['recency_diff'], bins=5)
```

```python
customer_df['recency_diff'].value_counts().sort_index()
```

```python
# recency는 최대최소를 기준으로 5구간으로 나눔
# recency_diff는 숫자가 클수록 오래 전에 방문했다는 뜻으로 등급이 낮아야 한다.
# 구간을 기준점으로 나누 것이 아니라 비율을 비슷하게 하는 qcut()으로 나눔

customer_df['r_score'] = pd.qcut(customer_df['recency_diff'], q=5, labels=['5', '4', '3', '2', '1'])
customer_df
```

```python
# 구간별 개수 확인
customer_df['r_score'].value_counts()
```

```python
# frequency는 구간 별로 개인적으로 설정값을 넣어줌. 
# 빈도 수에 따라 사람들의 분포가 차이가 많이 남. 그래서 설정값을 넣어 주는게 나을 것으로 생각

customer_df['f_score']= pd.cut(customer_df['frequency'], bins=[0, 1, 2, 3, 4, 34], labels=['1', '2', '3', '4', '5'])
customer_df
```

```python
customer_df['frequency'].value_counts()
```

```python
# 원래는 cut()함수를 썼는데 왜도가 심해서 5그룹에 해당하는 사람 수가 매우 적음 그래서 그냥 비슷한 비율로 나누는 qcut()함수 사용

customer_df['m_score']= pd.qcut(customer_df['monetary'], q=5, labels=['1', '2', '3', '4', '5'])
customer_df
```

```python
customer_df['m_score'].value_counts()
```

```python
# customer_df['RFM score'] = '(' + customer_df['r_score'].astype(str) + ',' + customer_df['f_score'].astype(str) + ',' + customer_df['m_score'].astype(str) + ')'
# customer_df
```

```python
customer_df['RFM_score'] = customer_df['r_score'].astype(str) + customer_df['f_score'].astype(str) + customer_df['m_score'].astype(str)
customer_df
```

```python
segments = {
    r'111|112|121|131|141|151' : '이탈 고객',
    r'332|322|233|232|223|222|132|123|122|212|211' : '동면 고객', 
    r'155|154|144|214|215|115|114|113' : '놓치면 안 될 고객',
    r'255|254|245|244|253|252|243|242|235|234|225|224|153|152|145|143|142|135|134|133|125|124' : '이탈 우려 고객',
    r'331|321|312|221|213|231|241|251' : '휴면 예정 고객',
    r'535|534|443|434|343|334|325|324' : '관심 필요 고객',
    r'525|524|523|515|514|513|425|424|413|414|415|315|314|313' : '잠재 고객',
    r'522|521|512|511|422|421|412|411|311' : '신규 고객',
    r'553|551|552|541|542|533|532|531|452|451|442|441|431|453|433|432|423|353|352|351|342|341|333|323' : '잠재 충성 고객',
    r'543|444|435|355|354|345|344|335' : '충성 고객',
    r'555|554|544|545|454|455|445' : 'VIP 고객'
}

customer_df['segment'] = customer_df['rfm_score'].replace(segments, regex=True)
```

```python
df_by_gender = pd.DataFrame(merged_df.groupby(['성별', '고객지역'])['고객ID'].count().reset_index(name='count'))
df_by_gender
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='고객지역', y='count', hue='성별', data=df_by_gender)

# 그래프 제목과 축 레이블 추가
plt.title('성별 및 고객지역에 따른 고객 수')
plt.xlabel('고객지역')
plt.ylabel('고객 수')

# 범례 추가
plt.legend(title='성별')

# 그래프 표시
plt.show()
```
