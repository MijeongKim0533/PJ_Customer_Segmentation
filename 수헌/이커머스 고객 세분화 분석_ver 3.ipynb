```python
df = pd.DataFrame(online_cus)


df['거래날짜'] = pd.to_datetime(df['거래날짜'])
df
```

# Recency

```python
# 고객 별로 가장 최근에 구매한 날짜 구하기
recent_purchase = df.groupby('고객ID')['거래날짜'].max()
recent_purchase
```

```python
df_sorted = df.sort_values(by='거래날짜', ascending=False)
df_sorted
```

```python
latest_date = df_sorted.iloc[0]['거래날짜']

# 'diff' 열 추가
df_sorted['date_diff'] = (latest_date - df_sorted['거래날짜']).dt.days
df_sorted
```

```python
df_sorted['R_score'] = pd.qcut(df_sorted['date_diff'], q=5, labels=[5, 4, 3, 2, 1])
df_sorted
```

# Frequency

```python
df
```

```python
frequency= df.groupby('고객ID')['거래날짜'].nunique().reset_index()
frequency
```

```python
frequency['거래날짜'].value_counts()
# 대부분이 1회, 2회만 구매한 고객들
```

```python
def assign_score(purchase_count):
    if purchase_count == 1:
        return 1
    elif purchase_count == 2:
        return 2
    elif purchase_count == 3:
        return 3
    elif purchase_count == 4:
        return 4
    elif purchase_count >= 5:
        return 5

# '점수' 열 추가
df['점수'] = frequency['거래날짜'].apply(assign_score)
df
```

```python
df['점수'].value_counts()
```
